## Step 6: K-Means from Scratch

K-Means works in 3 simple steps:
1. **Initialize** — randomly pick K points as starting cluster centers (centroids)
2. **Assign** — assign every point to its nearest centroid
3. **Update** — move each centroid to the average of all points assigned to it

Repeat steps 2 and 3 until centroids stop moving (convergence).

```python
def euclidean_distance(point1, point2):
    """
    Calculate straight-line distance between two points.
    point1 and point2 are arrays like [latitude, longitude]
    """
    distance = np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)
    return distance
```

```python
def assign_clusters(data, centroids):
    """
    For every point in data, find which centroid it is closest to.
    Returns a list of cluster labels (0, 1, 2 ... K-1)
    """
    labels = []

    for point in data:
        distances = []
        for centroid in centroids:
            d = euclidean_distance(point, centroid)
            distances.append(d)

        # Pick the centroid with smallest distance
        closest = distances.index(min(distances))
        labels.append(closest)

    return labels
```

```python
def update_centroids(data, labels, K):
    """
    Recalculate centroid positions as the mean of all points in each cluster.
    """
    new_centroids = []

    for k in range(K):
        # Collect all points assigned to cluster k
        cluster_points = [data[i] for i in range(len(data)) if labels[i] == k]

        if len(cluster_points) == 0:
            # Edge case: if a cluster has no points, keep old centroid
            # (this rarely happens but just to be safe)
            new_centroids.append(data[np.random.randint(0, len(data))])
        else:
            # Calculate the mean position of all points in this cluster
            mean_lat = sum(p[0] for p in cluster_points) / len(cluster_points)
            mean_lon = sum(p[1] for p in cluster_points) / len(cluster_points)
            new_centroids.append([mean_lat, mean_lon])

    return new_centroids
```

```python
def kmeans(data, K, max_iterations=100, random_seed=42):
    """
    Full K-Means algorithm from scratch.

    Parameters:
        data           : list of [lat, lon] points
        K              : number of clusters we want
        max_iterations : stop after this many iterations even if not converged
        random_seed    : for reproducibility

    Returns:
        labels    : cluster assignment for each point
        centroids : final centroid positions
    """
    np.random.seed(random_seed)

    # Step 1: Randomly pick K points from data as initial centroids
    random_indices = np.random.choice(len(data), K, replace=False)
    centroids = [data[i] for i in random_indices]

    labels = [0] * len(data)  # start with all labels as 0

    for iteration in range(max_iterations):
        # Step 2: Assign each point to closest centroid
        new_labels = assign_clusters(data, centroids)

        # Step 3: Update centroids
        new_centroids = update_centroids(data, new_labels, K)

        # Check for convergence: if labels didn't change at all, we're done!
        if new_labels == labels:
            print(f"Converged after {iteration + 1} iterations!")
            break

        labels = new_labels
        centroids = new_centroids

    return labels, centroids
```

## Step 7: Prepare Data and Run K-Means

```python
# Extract latitude and longitude as a list of [lat, lon] pairs
data_points = telangana_df[["Latitude", "Longitude"]].values.tolist()

print("Total data points:", len(data_points))
print("First 3 points:", data_points[:3])
```

```python
# Choose K = 6 clusters (Maharashtra has 6 major geographical divisions:
# Konkan, Western Maharashtra, Marathwada, Vidarbha, North Maharashtra, Mumbai Metro)
# You can change this number and re-run to experiment!
K = 6

print(f"Running K-Means with K = {K} ...")
labels, centroids = kmeans(data_points, K=K, max_iterations=100, random_seed=42)

print(f"\nClustering done!")
print(f"Cluster labels (first 10): {labels[:10]}")
print(f"\nFinal Centroids (Lat, Lon):")
for i, c in enumerate(centroids):
    print(f"  Cluster {i}: Latitude = {c[0]:.4f}, Longitude = {c[1]:.4f}")
```

## Step 8: Add Cluster Labels Back to Dataframe

```python
# Add the cluster labels as a new column
telangana_df["Cluster"] = labels

# Check how many pincodes are in each cluster
print("Pincode count per cluster:")
print(telangana_df["Cluster"].value_counts().sort_index())
```

## Step 9: Visualize the Clusters

```python
# Plot each cluster in a different color
colors = ["crimson", "steelblue", "forestgreen", "darkorange", "purple",
          "brown", "deeppink", "teal", "goldenrod", "navy"]

plt.figure(figsize=(12, 9))

for k in range(K):
    cluster_data = telangana_df[telangana_df["Cluster"] == k]
    plt.scatter(
        cluster_data["Longitude"],
        cluster_data["Latitude"],
        s=8,
        color=colors[k],
        alpha=0.6,
        label=f"Cluster {k} ({len(cluster_data)} pincodes)"
    )

# Plot the centroids as big stars
for i, c in enumerate(centroids):
    plt.scatter(
        c[1], c[0],          # note: lon is x-axis, lat is y-axis
        s=250,
        color=colors[i],
        marker="*",
        edgecolors="black",
        linewidths=1.2,
        zorder=5,
        label=f"Centroid {i}"
    )

plt.title(f"K-Means Clustering of Maharashtra Pincodes (K={K})", fontsize=15, fontweight="bold")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(loc="lower right", fontsize=8, markerscale=1.5)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Step 10: Choosing the Right K — Elbow Method
We can use the **Within-Cluster Sum of Squares (WCSS)** to pick a good value of K.
WCSS = total squared distance of every point from its centroid.
As K increases, WCSS drops. We look for the "elbow" where it stops dropping sharply.

```python
def calculate_wcss(data, labels, centroids, K):
    """
    Calculate Within-Cluster Sum of Squares.
    Lower = tighter, better clusters.
    """
    wcss = 0
    for i in range(len(data)):
        centroid = centroids[labels[i]]
        dist = euclidean_distance(data[i], centroid)
        wcss += dist ** 2
    return wcss
```

```python
# Try K values from 1 to 9 and record WCSS
# Note: this takes a bit of time since we wrote K-Means from scratch!
wcss_values = []
k_range = range(1, 10)

for k in k_range:
    print(f"Running K-Means for K={k}...", end=" ")
    lbs, cents = kmeans(data_points, K=k, max_iterations=50, random_seed=42)
    wcss = calculate_wcss(data_points, lbs, cents, k)
    wcss_values.append(wcss)
    print(f"WCSS = {wcss:.4f}")
```

```python
# Plot the Elbow Curve
plt.figure(figsize=(9, 5))
plt.plot(list(k_range), wcss_values, marker="o", color="steelblue", linewidth=2, markersize=8)
plt.title("Elbow Method — Finding the Best K", fontsize=14, fontweight="bold")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.xticks(list(k_range))
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

print("Look for the 'elbow' — where the curve starts flattening out.")
print("That K value is usually the best choice!")
```
